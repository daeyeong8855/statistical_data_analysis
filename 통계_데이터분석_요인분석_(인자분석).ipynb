# 인자 개수 결정법
* Scree plot
* 카이저 기준 (Kaiser criterion)

* 고유값 : 중요도
* 고유벡터 : 방향

** 이후 구조방정식 모델링으로 이어짐

```python
import numpy as np
from sklearn.decomposition import FactorAnalysis

# 데이터 생성
n_samples = 100 #행
n_feature = 5   #열

# 데이터셋을 랜덤하게 설정
x = np.random.randn(n_samples, n_feature)

# 요인분석 수행
n_components =2 # 추출할 인자 수
factorAnalysis = FactorAnalysis(n_components=n_components, random_state=0) # 시드고정
x_factor = factorAnalysis.fit_transform(x)

# 결과 출력
print(f"원본 데이터 크기 {x.shape}")
print(f"인자 분석 결과 인자 개수 : {n_components}")
print(x_factor)
```
