```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10
num_simulations = 1000
t_values = np.random.standard_t(df, num_simulations)

plt.hist(t_values, bins=20, alpha=0.5)
plt.show()
```
